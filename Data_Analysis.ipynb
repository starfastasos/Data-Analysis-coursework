**Οπτικοποίηση των δεδομένων**

Στο ακόλουθο τμήμα κώδικα απεικονίζουμε σε ένα κυκλικό διάγραμμα τα ποσοστά των δύο τιμών που παίρνει το χαρακτηριστικό - στόχος. Ουσιαστικά εννοούμε τι ποσοστό του συνόλου δεδομένων έδωσε τη τιμή 0 και τι τη τιμή 1. Όπως παρατηρούμε το 46.2% απάντησε τη τιμή 0 και το υπόλοιπο 53.8% τη τιμή 1.

```python
#Code that uses the Matplotlib library to draw a pie chart based on the values in column 'D' of a DataFrame named 'Data'

# Importing the required library
import matplotlib.pyplot as plt

# Function to plot a pie chart based on column 'D' of DataFrame 'Data'
def plot_pie_chart(Data, col):
    # Counting the values in column 'D'
    counts = Data[col].value_counts()
    
    # Extracting the unique values and their counts
    labels = counts.index
    values = counts.values
    
    # Creating a new figure
    plt.figure()
    
    # Plotting the pie chart with labels and autopct formatting
    plt.pie(values, labels=labels, autopct='%1.1f%%')
    
    # Setting the title of the pie chart
    plt.title("Pie Chart of Feature {}".format(col))

# Calling the function to plot the pie chart using column 'D' of DataFrame 'Data'
plot_pie_chart(Data, 'D')
```

**Ακραίες Τιμές**

Τα χαρακτηριστικά που περιέχει το σύνολο δεδομένων μας είναι σε κλίμακα Likert, δηλαδή από το 1 έως το 5. Αυτό δεν καθιστά αξιόπιστο ότι οι ακραίες τιμές του θα είναι οι τιμές 1 και 5. Συνεπώς, η έννοια της ακραίας τιμής για το συγκεκριμένο σύνολο δεν έχει καίρια σημασία.

**Μετασχηματισμός Μεταβλητών**

Στο δεδομένα μας θα χρησιμοποιήσουμε τη μέθοδο της τυποποίησης. Για το λόγο αυτό θα χρησιμοποιήσουμε τη συνάρτηση StandardScaler() της scikit-learn. Η μέθοδος της τυποποίησης μετασχηματίζει τα δεδομένα μίας μεταβλητής έτσι ώστε να διαθέτουν μέση τιμή 0 και τυπική απόκλιση 1.

```python
# Importing the required library for standardizing variables
from sklearn.preprocessing import StandardScaler

# Standardizing the variables in the 'Data' DataFrame
# The 'fit_transform' method scales the values of each column
scaled_columns = StandardScaler().fit_transform(Data.values)

# Creating a new DataFrame 'scaled_dt' using the standardized values
# The index and column names are preserved from the original 'Data' DataFrame
scaled_dt = pd.DataFrame(scaled_columns, index=Data.index, columns=Data.columns)
```

**Δημιουργία συνόλων εκπαίδευσης και ελέγχου**

Ένας απο τους καθοριστικούς παράγοντες για την εκτίμηση των δυνατοτήτων ενός αλγορίθμου μηχανικής μάθησης είναι ο ορισμός του συνόλου εκπαίδευσης που θα χρησιμοποιηθεί για να εκπαιδευτεί ο αλγόριθμος ώστε να ανταποκρίνεται με σωστό τρόπο σε μελλοντικά δεδομένα (αποτελεσματικότητα στην πραγματοποίηση ορθών προβλέψεων) καθώς και ο ορισμός του συνόλου ελέγχου που θα χρησιμοποιηθεί ώστε να ελεγχθεί η συγκεκριμένη αποτελεσματικότητα. Για το λόγο αυτό θα χρησιμοποιήσουμε τη συνάρτηση train_test_split() της scikit-learn. Στη συνάρτηση αυτή περνάμε μία συγκεκριμένη τιμή στην παράμετρο random_state (δηλαδή τη τιμή 42) προκειμένου κάθε φορά που εκτελώ το πρόγραμμα να γίνεται ο διαχωρισμός του συνόλου δεδομένων σε δεδομένα εκπαίδευσης και δεδομένα ελέγχου με τον ίδιο ακριβώς τρόπο. Το σύνολο εκπαίδευσης είναι το 60% και το σύνολο ελέγχου το 40% του συνόλου δεδομένων, αυτό το καθορίζουμε με τη τιμη 0.4 στη μεταβλητή test_size.

```python
# Importing the required library for creating training and test sets
from sklearn.model_selection import train_test_split

# Splitting the dataset 'scaled_dt' into training and test sets
# The test set will have 40% of the data, and the random_state ensures reproducibility of the split
train_set, test_set = train_test_split(scaled_dt, test_size=0.4, random_state=42)

# Extracting the predictors from the training set by dropping the 'D' column
predictors_train = train_set.drop('D', axis=1)

# Extracting the labels from the original 'Data' DataFrame based on the indices of the predictors in the training set
labels_train = Data.loc[predictors_train.index, 'D']

# Extracting the predictors from the test set by dropping the 'D' column
predictors_test = test_set.drop('D', axis=1)

# Extracting the labels from the original 'Data' DataFrame based on the indices of the predictors in the test set
labels_test = Data.loc[predictors_test.index, 'D']
```

**1. Δέντρα απόφασης (Decisions Trees):**

Τα δέντρα αποφάσης είναι μια δημοφιλής τεχνική μηχανικής μάθησης που χρησιμοποιείται ευρέως για προβλήματα κατηγοριοποίησης. Η μέθοδος αυτή περιλαμβάνει την αναδρομική κατάτμηση του χώρου χαρακτηριστικών σε όλο και μικρότερες περιοχές, με βάση τα πιο κατατοπιστικά χαρακτηριστικά, μέχρι να ικανοποιηθεί ένα κριτήριο διακοπής. Τα δέντρα αποφάσης είναι εύκολο να ερμηνευθούν και να απεικονιστούν, αλλά είναι επιρρεπή στην υπερπροσαρμογή και μπορεί να είναι ευαίσθητα σε μικρές αλλαγές στα δεδομένα εκπαίδευσης. Η χρήση τους στο σύνολο δεδομένων μας γίνεται γιατί αυτά μπορούν να συλλάβουν μη γραμμικές σχέσεις μεταξύ των χαρακτηριστικών και της μεταβλητής στόχου.

```python
# Importing the required class for the Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# Creating an instance of the Decision Tree Classifier
dec_tree = DecisionTreeClassifier()

# Fitting the classifier to the training data
dec_tree.fit(predictors_train, labels_train)
```

**2. Μηχανές Διανυσμάτων Υποστήριξης (Support Vector Machines):**

Οι μηχανές διανυσμάτων υποστήριξης είναι ένας αλγόριθμος μηχανικής μάθησης που χρησιμοποιείται για προβλήματα κατηγοριοποίησης δυαδικών και πολλαπλών κλάσεων. Ο στόχος τους είναι να βρουν ένα υπερεπίπεδο που διαχωρίζει στο μέγιστο βαθμό τις κλάσεις στο χώρο χαρακτηριστικών. Οι μηχανές αυτές το επιτυγχάνουν βρίσκοντας το υπερεπίπεδο με το μέγιστο περιθώριο μεταξύ των κλάσεων, το οποίο είναι η απόσταση μεταξύ του υπερεπιπέδου και των πλησιέστερων σημείων δεδομένων από κάθε κλάση. Επιπλέον, μπορούν να χειριστούν τόσο γραμμικά όσο και μη γραμμικά όρια απόφασης χρησιμοποιώντας συναρτήσεις πυρήνα και είναι λιγότερο επιρρεπής στην υπερπροσαρμογή από τα δέντρα απόφασης, επομένως θα είναι για το σύνολό μας και κυρίως το σύνολο δοκιμής μας περισσότερο αξιόπιστες. Ωστόσο, μπορούν να είναι ευαίσθητες στην επιλογή της συνάρτησης πυρήνα και των υπερπαραμέτρων, και μπορεί να είναι υπολογιστικά δαπανηρές για μεγάλα σύνολα δεδομένων.

```python
# Importing the required class for the Support Vector Classifier
from sklearn.svm import SVC

# Creating an instance of the Support Vector Classifier
svm = SVC(probability=True)

# Fitting the classifier to the training data
svm.fit(predictors_train, labels_train)
```

Εκτιμά τη σχέση μεταξύ των μεταβλητών εισόδου και του αποτελέσματος χρησιμοποιώντας μια λογιστική συνάρτηση, η οποία αντιστοιχίζει κάθε είσοδο σε μια τιμή πιθανότητας μεταξύ 0 και 1. Η λογιστική παλινδρόμηση είναι εύκολη στην ερμηνεία και παρέχει πιθανοτικές προβλέψεις, αλλά υποθέτει ότι οι μεταβλητές εισόδου έχουν γραμμική σχέση με τις λογαριθμικές πιθανότητες του δυαδικού αποτελέσματος. Ακόμη, μπορεί να είναι ευαίσθητη σε ακραίες τιμές και δεδομένα υψηλής διάστασης κάτι που δεν επηρεάζει το σύνολό μας. Συμπερασματικά, είναι μία καλή μέθοδος κατηγοριοποίησης για το σύνολο δεδομένων μας.

```python
# Importing the required class for the Logistic Regression
from sklearn.linear_model import LogisticRegression

# Creating an instance of the Logistic Regression
log_reg = LogisticRegression()

# Fitting the classifier to the training data
log_reg.fit(predictors_train, labels_train)
```

### **Μέθοδοι Αξιολόγησης**

**Ακρίβεια (Accuracy):**

Η ακρίβεια είναι η πιο κοινή μετρική που χρησιμοποιείται για την αξιολόγηση των μοντέλων κατηγοριοποίησης και μετρά το ποσοστό των σωστά ταξινομημένων περιπτώσεων επί του συνόλου των περιπτώσεων. Ωστόσο, η μετρική αυτή μπορεί να είναι παραπλανητική εάν οι κλάσεις είναι ανισόρροπες και δεν αποτυπώνει το κόστος των λαθών λανθασμένης ταξινόμησης.

**F1 - Score:**

Το F1-score είναι ένας σταθμισμένος μέσος όρος της ακρίβειας και της ανάκλησης και εξισορροπεί το trade-off μεταξύ των δύο μετρικών. Αυτός ο μετρητής αξιολόγησης είναι χρήσιμος όταν τόσο τα ψευδώς θετικά όσο και τα ψευδώς αρνητικά είναι σημαντικά.

**Καμπύλη ROC:**

Η καμπύλη ROC είναι μια γραφική αναπαράσταση του αληθούς θετικού ποσοστού  έναντι του ψευδώς θετικού ποσοστού καθώς μεταβάλλεται το κατώφλι ταξινόμησης. Η περιοχή κάτω από την καμπύλη ROC (AUC) μετρά τη συνολική απόδοση του κατηγοριοποιητή και είναι χρήσιμη όταν το κόστος των ψευδώς θετικών και των ψευδώς αρνητικών δεν είναι ίσο. Η AUC κυμαίνεται από 0.5 (τυχαίος κατηγοριοποιητής) έως 1 (τέλεια κατηγοριοποίηση).

### **Εφαρμογή Μεθόδων Αξιολόγησης και Συμπεράσματα**

```python
# Import the necessary modules and functions
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import auc

# Define the parameter grid for SVM tuning
svm_param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Create the SVM classifier
svm = SVC(probability=True)

# Perform grid search with cross-validation for SVM
svm_grid_search = GridSearchCV(svm, svm_param_grid, cv=5, scoring='accuracy')
svm_grid_search.fit(predictors_train, labels_train)

# Print the best parameters and the corresponding accuracy score for SVM
print("Accuracy for SVM: ", svm_grid_search.best_score_)

# Get the best SVM classifier with the optimized hyperparameters
best_svm = svm_grid_search.best_estimator_

# Define the parameter grid for Logistic Regression tuning
lr_param_grid = {
    'C': [0.1, 1, 10],
    'solver': ['liblinear', 'saga']
}

# Create the Logistic Regression classifier
lr = LogisticRegression()

# Perform grid search with cross-validation for Logistic Regression
lr_grid_search = GridSearchCV(lr, lr_param_grid, cv=5, scoring='accuracy')
lr_grid_search.fit(predictors_train, labels_train)

# Print the best parameters and the corresponding accuracy score for Logistic Regression
print("Accuracy for Logistic Regression: ", lr_grid_search.best_score_)

# Get the best Logistic Regression classifier with the optimized hyperparameters
best_lr = lr_grid_search.best_estimator_

# Define the parameter grid for Naive Bayes tuning
nb_param_grid = {}

# Create the Naive Bayes classifier
nb = GaussianNB()

# Perform grid search with cross-validation for Naive Bayes
nb_grid_search = GridSearchCV(nb, nb_param_grid, cv=5, scoring='accuracy')
nb_grid_search.fit(predictors_train, labels_train)

# Print the best parameters and the corresponding accuracy score for Naive Bayes
print("Accuracy for Naive Bayes: ", nb_grid_search.best_score_)

# Get the best Naive Bayes classifier with the optimized hyperparameters
best_nb = nb_grid_search.best_estimator_


# Predict labels for the test data using each classifier
svm_labels_pred = best_svm.predict(predictors_test)
lr_labels_pred = best_lr.predict(predictors_test)
nb_labels_pred = best_nb.predict(predictors_test)

# Calculate the F1 score for each classifier
svm_f1 = f1_score(labels_test, svm_labels_pred)
lr_f1 = f1_score(labels_test, lr_labels_pred)
nb_f1 = f1_score(labels_test, nb_labels_pred)

print()
print("F1-Score for SVM: ", svm_f1)
print("F1-Score for Logistic Regression: ", lr_f1)
print("F1-Score for Naive Bayes: ", nb_f1)
print()

# Calculate the probabilities for the positive class using each classifier
svm_probs = best_svm.predict_proba(predictors_test)[:, 1]
lr_probs = best_lr.predict_proba(predictors_test)[:, 1]
nb_probs = best_nb.predict_proba(predictors_test)[:, 1]

#Compute the false positive rate, true positive rate, and threshold for ROC curve for each classifier
svm_fpr, svm_tpr, svm_thresholds = roc_curve(labels_test, svm_probs)
lr_fpr, lr_tpr, lr_thresholds = roc_curve(labels_test, lr_probs)
nb_fpr, nb_tpr, nb_thresholds = roc_curve(labels_test, nb_probs)

#Compute the area under the ROC curve for each classifier
svm_roc_auc = auc(svm_fpr, svm_tpr)
lr_roc_auc = auc(lr_fpr, lr_tpr)
nb_roc_auc = auc(nb_fpr, nb_tpr)

print("ROC AUC for SVM: ", svm_roc_auc)
print("ROC AUC for Logistic Regression: ", lr_roc_auc)
print("ROC AUC for Naive Bayes: ", nb_roc_auc)

#Plot the ROC curve for each classifier
plt.plot(svm_fpr, svm_tpr, label='SVM (AUC = %0.2f)' % svm_roc_auc)
plt.plot(lr_fpr, lr_tpr, label='Logistic Regression (AUC = %0.2f)' % lr_roc_auc)
plt.plot(nb_fpr, nb_tpr, label='Naive Bayes (AUC = %0.2f)' % nb_roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```
